```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdata = np.arange(.1,10,.5)
ydata = xdata**2
yerror = ydata**-.3
plt.figure(1)
plt.errorbar(xdata, ydata, yerr=yerror)
#This is the raw call to errorbar(). Notice that it give syou the data as well as the error bars.

plt.figure(2)
plt.errorbar(xdata, ydata, yerr=yerror*10, capsize=5, color = 'k', marker = '.', linestyle = 'None')
plt.xticks((0,3,4,7,10)) #I don't know why you'd want tick marks there, but here's how to manually set tick marks
plt.xlabel("$\lambda_{max}$ (nm$^2$)") #special characters, suscripts, and superscripts
plt.ylabel("Bigger font!", fontsize=20)
plt.legend(['a curve with error bars'], fontsize=14, loc='upper right')
#That legend is in a dumb spot to show you how to manually place it. The default tries to place it out of the way

plt.savefig("error bars.png") #this saves the current figure to an image file. You can do several file formats.
```
